1.1 Line Plot

A line graph is the simplest plot that displays the relationship between the one independent and one dependent dataset.In this plot, the points are joined by straight line segemnts.

```python
import numpy as np
X = np.linspace(1,20,100)   #(a,b,c) 表示，从a到b中取样c次
Y = np.exp(X)   #y=e的x次幂

#line plot
plt.plot(X,Y)

plt.show()
```

```python
X = np.linspace(1,20,100)
Y = np.exp(X)

plt.plot(X,Y,'g*')  #'g*' 代表用绿色*画图，同理r*,b*
plt.show()
```

```python
X = np.linspace(1,20,100)
Y1 = X  #y=x
Y2 = np.square(X)   #y=x的平方
Y3 = np.sqrt(X)     #y=根号x


plt.plot(X,Y1,'r',X,Y2,'b',X,Y3,'g')

#add title to the plot
plt.title('Line Plot')
plt.show()
```

1.2 Scatter Plot


A scatter plot is a set of points plotted on horizontal an vertical axes. The Scatter plot can be used to study the correlation between the two variables. One can also detect the extreme data points using a scatter plot.

散点图是一组绘制在x轴和y轴上的点。散点图可以用来研究两个变量之间的相关性。我们还可以使用散点图来检测极端数据点。

```python
tips_data.head()
```

```python
X = tips_data['total_bill']
Y = tips_data['tip']

plt.scatter(X,Y)

#   x和y的标签填写
plt.xlabel('total_bill')
plt.ylabel('tip')

plt.show()
```

```python
X = tips_data['total_bill']
Y = tips_data['tip']

#alpha 透明度 0到1
#s是标记点大小，默认20，可以是数字也可以是数组，**代表幂运算
plt.scatter(X,Y, s=np.array(Y)**1.5, c='blue', alpha=0.5) #随着y的增大，标记点的大小也在增大

plt.title('Scatter Plot')
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.show()
```

1.3 Bar Plot

A bar plot is used to display categorical data with bars
条形图用于显示分类数据，条形的长度与它们所代表的值成正比。类别型变量的不同类别之间的比较可以通过研究条形图来完成。在垂直条形图中，x轴显示分类变量，y轴包含不同类别对应的值。

```python
tips_data.head()
```

```python
# the variable 'smoker is categorical  Smoker是分类变量
# check categories in the variable     检查smoker标签中的变量都有什么
set(tips_data['smoker'])
#set(tips_data['day']) 可以看出来对比
```

```python
tips_data.smoker.value_counts().plot(kind='bar', rot=0, color='green')


#x,y数字在坐标轴上的位置
#s is the text to be added
#因为tips_data.smoker.value_counts()[]这个数组只有两个数据，一个是yes的数字，一个是no的数字
plt.text(x=0.98, y=tips_data.smoker.value_counts()[1]+1, s=tips_data.smoker.value_counts()[1])
plt.text(x=-0.05, y=tips_data.smoker.value_counts()[0]+2, s=tips_data.smoker.value_counts()[0])

plt.title('Bar Plot')
plt.xlabel('Smoker')
plt.ylabel('Count')
plt.show()
tips_data.smoker.value_counts()
```

1.6 Box Plot

箱线图是一种可视化变量的五位数摘要的方法。五位数总结包括最小值、第一四分位数(Q1)、中位数(Q2)、第三四分位数(Q3)和最大值等数值。箱线图给出了数据中异常值的信息。检测和去除异常值是探索性数据分析中最重要的步骤之一。箱形图还说明了数据的分布情况。


```python
tips_data.head()
```

```python
plt.boxplot(tips_data['total_bill'])

#add labels for five number summary
plt.text(x=1.1, y=tips_data['total_bill'].min(), s='min')
plt.text(x=1.1, y=tips_data.total_bill.quantile(0.25).min(), s='Q1') # 四分之一位数
plt.text(x=1.1, y=tips_data['total_bill'].median(), s='median Q2')
plt.text(x=1.1, y=tips_data.total_bill.quantile(0.75), s='Q3') # 四分之三位数
plt.text(x=1.1, y=tips_data['total_bill'].max(), s='max')

plt.ylabel('Total Bill')
plt.title('Boxplot of Total Bill Amount')
plt.show()
```

```python
#现在我们要加入一些文字来表示脱离正常值的特殊值
plt.boxplot(tips_data['total_bill'])

#add labels for five number summary
plt.text(x=1.1, y=tips_data['total_bill'].min(), s='min')
plt.text(x=1.1, y=tips_data.total_bill.quantile(0.25).min(), s='Q1') # 四分之一位数
plt.text(x=1.1, y=tips_data['total_bill'].median(), s='median Q2')
plt.text(x=1.1, y=tips_data.total_bill.quantile(0.75), s='Q3') # 四分之三位数
plt.text(x=1.1, y=tips_data['total_bill'].max(), s='max')


#xy 为被注释的坐标点
#xytext 为注释文字的坐标位置

#arrowprops #箭头参数,参数类型为字典dict
    # width：箭头的宽度(以点为单位)
    # headwidth：箭头底部以点为单位的宽度
    # headlength：箭头的长度(以点为单位)
    # shrink：总长度的一部分，从两端“收缩”
    # facecolor：箭头颜色
plt.annotate(text='Outliers', xy = (0.97,45),xytext=(0.7, 45), arrowprops = dict(facecolor='black', arrowstyle = 'simple'))


plt.ylabel('Total Bill')
plt.title('Boxplot of Total Bill Amount')
plt.show()
```

2.1 Strip Plot

The strip plot resembles a scatterplot when one variable is categorical. This plot can help study the underlying distribution.
当一个变量是类别型变量时，条形图类似于散点图


```python
tips_data.head()
```

```python
#从Lunch和Dinner中查询数据，data在tips_data 中
sns.stripplot(y= 'tip',x='time', order=['Lunch','Dinner'], data=tips_data)

plt.show()
```

```python
#jitter的作用是将点分散开
sns.stripplot(y = 'tip', x = 'time', order = ['Lunch','Dinner'], data = tips_data, jitter = True)

plt.show()
```

2.2 Swarm Plot

Swarm的好处是，所有的数据点都不会重叠

```python
tips_data.head()
```

```python
sns.swarmplot(y='tip', x='time', data=tips_data)
plt.show()
```

```python
#add categorical variable in the above plot by using a parameter 'hue'
sns.swarmplot(y = 'tip', x = 'time', data = tips_data, hue = 'day')     # 同理可用'sex, smoker'等
plt.show()
```

2.3 Violin Plot

The violin plot is similar to a box plot that features a kernel density estimation of the underlying distribution. The plot shows the distribution of numerical variables across categories of one (or more) categorical variables such that those distributions can be compared.

小提琴图类似于箱线图，其特点是对基础分布的核密度估计。该图显示了一个(或多个)类别型变量的数值变量分布，以便可以比较这些分布。

```python
tips_data.head()
```

```python
sns.violinplot(y='total_bill', x='day', data=tips_data)
plt.show()
```

```python
# set the figure size 图大小
plt.figure(figsize = (10,5))

# violin plot with addition of the variable 'sex'
# 'split = True' draws half plot for each of the category of 'sex' 如果是False就分开画
sns.violinplot(y = 'total_bill', x = 'day', data = tips_data, hue = 'sex', split = True)

# display the plot
plt.show()
```

2.4 Pair Plot

The pair plot gives a pairwise distribution of variables in the dataset. pairplot() function creates a matrix such that each grid shows the relationship between a pair of variables. On the diagonal axes, a plot shows the univariate distribution of each variable.

配对图给出了数据集中变量的成对分布。Pairplot()函数创建一个矩阵，使每个网格显示一对变量之间的关系。在对角轴上，一个图显示了每个变量的单变量分布。

```python
tips_data.head()
```

```python
# set the figure size
plt.figure(figsize = (8,8))

# plot a pair plot
sns.pairplot(tips_data)

# display the plot
plt.show()
```

2.7 Heatmap

Heatmap is a two-dimensional graphical representation of data where the individual values that are contained in a matrix are represented as colors. Each square in the heatmap shows the correlation between variables on each axis. Correlation is a statistic that measures the degree to which two variables move with each other

热图是数据的二维图形表示，其中矩阵中包含的单个值用颜色表示。热图中的每个正方形都显示了每个轴上变量之间的相关性。相关性是一种度量两个变量相互移动程度的统计量

```python
tips_data.head()
```

Compute correlation between the variables using .corr() function. Plot a heatmap of the correlation matrix.

```python
corr_matrix = tips_data.corr()
corr_matrix
```

```python
sns.heatmap(corr_matrix,annot = True)  #annot 可以在图上输出数字
plt.show()
```
